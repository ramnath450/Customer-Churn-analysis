```python
plt.figure(figsize=(4,4))
plt.title("Churn by Gender", color="red")
ax=sns.countplot(df,
            x="gender",
            hue="Churn")
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
plt.figure(figsize=(4,4))
plt.title("Churn by SeniorCitizen", color="yellow")
ax =sns.countplot(df, x="SeniorCitizen", hue="Churn")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
churn_counts = df.groupby(['SeniorCitizen', 'Churn']).size().unstack(fill_value=0)

# Calculate the percentage
churn_percentages = churn_counts.div(churn_counts.sum(axis=1), axis=0) * 100

# Create the stacked bar chart
churn_percentages.plot(kind='bar', stacked=True, color=['lightblue', 'salmon'], figsize=(6, 4))

# Add labels and title
plt.title("Churn by Senior Citizen Status (%)", color="yellow")
plt.xlabel("Senior Citizen")
plt.ylabel("Percentage")
plt.xticks(ticks=[0, 1], labels=["No", "Yes"], rotation=0)
plt.legend(title="Churn", labels=["Not Churned", "Churned"])
plt.grid(axis='y')

# Display percentage labels on bars
for i in range(len(churn_percentages)):
    for j in range(len(churn_percentages.columns)):
        plt.text(i, churn_percentages.iloc[i, j] + 1, f"{churn_percentages.iloc[i, j]:.1f}%", ha='center')
plt.show()
```

```python
plt.title("Count of Customers by Contracts")
sns.set(rc={"figure.figsize":(4,4)})
ax = sns.countplot(df, x="Contract", hue="Churn")
for bars in ax.containers:
    ax.bar_label(bars)
```

People who have Month-to-Month contract are likely to be churn then those who have 1 or 2 year contract or more

```python
columns = ['PhoneService', 'MultipleLines', 'InternetService',
           'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 
           'TechSupport', 'StreamingTV', 'StreamingMovies']

# Set up the matplotlib figure with subplots
fig, axes = plt.subplots(3, 3, figsize=(10, 10))  # 3x3 grid for 9 columns
fig.suptitle('Subplots for Categorical Features', fontsize=16)

# Loop through the columns and axes to plot count plots
for i, column in enumerate(columns):
    ax = axes[i//3, i%3]  # Identify subplot row/column index
    sns.countplot(x=df[column], ax=ax, hue=df["Churn"])  # Create count plot
    ax.set_title("CountPlot of"+ column)  # Set title

plt.tight_layout()
plt.show()
```

For most services like 'PhoneService', 'InternetService', and 'StreamingTV', a higher number of non-churned customers is evident. Certain services like 'Fiber optic' in 'InternetService' and 'No' in 'OnlineSecurity' and 'TechSupport' seem to have a higher association with churn, as indicated by the relative increase in churn for these categories.
